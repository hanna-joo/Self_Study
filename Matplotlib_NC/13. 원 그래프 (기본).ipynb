# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window
# matplotlib.rcParams['font.family'] = 'KoPubBatang_Pro' # Apple
matplotlib.rcParams['font.family'] = 'AppleGothic' # Apple
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
values = [1, 1, 1, 1, 1, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f')
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f')
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90) # 시작점을 90도 왼쪽으로 돌려서 그래프 그리기
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) # 시계방향으로 90도 돌려서 그래프 그리기
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.2, 0.1, 0, 0, 0, 0] # 그래프에서 거리를 둘 만큼의 크기 개별로 설정

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.05] * 6 # 그래프에서 거리를 둘 만큼의 크기를 동일하게 설정

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```
