# 9. 막대 그래프 (심화)

- [hatch 참고](https://matplotlib.org/devdocs/gallery/shapes_and_collections/hatch_style_reference.html)

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window
# matplotlib.rcParams['font.family'] = 'KoPubBatang_Pro' # Apple
matplotlib.rcParams['font.family'] = 'AppleGothic' # Apple
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.barh(labels, values) # 수평으로 뻗어나가는 막대그래프
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ...
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):  # rect는 각 bar에 대한 위치, 크기 정보
    plt.text(idx, rect.get_height(), values[idx]) # x좌표는 idx, y좌표는 bar의 높이, 표현할 텍스트 값
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):  # rect는 각 bar에 대한 위치, 크기 정보
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='blue')
```
